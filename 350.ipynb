```python
from PIL import Image
from ultralytics import YOLO
import os
import pyttsx3

# Load a pretrained YOLOv8n model
model = YOLO('D:/FIRE_FLAME/nano_350/detect/train/weights/best.pt')

# Folder
image_files = [f for f in os.listdir('D:/FIRE_FLAME/data/images/test') ]

# Run inference on 'bus.jpg'
index=176
# results = model('D:/FIRE_FLAME/data/images/test'+'/'+image_files[index])  # results list
results = model('test1.png')  # results list

#speak for fire
x=""
for r in results:
    x=r.boxes.conf
x=str(x)[8:str(x).rfind(']')]  

def speak(text):
    engine = pyttsx3.init()
    engine.setProperty('rate', 190)
    engine.say(text)
    engine.runAndWait()

# Show the results
for r in results:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image
    im.save('results.jpg')  # save image
    if len(x)>1:
            speak("Fire detected ")
```

```python
import supervision as sv
import numpy as np
from ultralytics import YOLO

VIDEO_PATH = "test.mp4"

model = YOLO("best.pt")

video_info = sv.VideoInfo.from_video_path(VIDEO_PATH)
```

```python
def process_frame(frame: np.ndarray, _) -> np.ndarray:
    results = model(frame, imgsz=1280)[0]
    
    detections = sv.Detections.from_yolov8(results)

    box_annotator = sv.BoxAnnotator(thickness=4, text_thickness=4, text_scale=2)

    labels = [f"{model.names[class_id]} {confidence:0.2f}" for _, _, confidence, class_id, _ in detections]
    frame = box_annotator.annotate(scene=frame, detections=detections, labels=labels)

    return frame

sv.process_video(source_path=VIDEO_PATH, target_path=f"result.mp4", callback=process_frame)
```

**PREDICTION for Video Input**

```python
import cv2
from ultralytics import YOLO
import os

video_path = os.path.join(r"D:\animal detection\elephant-bear\model\runs\detect\predict\bear.mp4") ## input


video_path_out = '{}_out.mp4'.format(video_path)

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

# Load a model
model = YOLO(r'D:\animal detection\elephant-bear\model\runs\detect\train5\weights\best.pt')  # load a custom model

threshold = 0.5

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
